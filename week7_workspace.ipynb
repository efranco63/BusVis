```python
from datetime import datetime
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import cm as cm
%matplotlib inline
import scipy.interpolate
import pandas as pd
```

```python
#boundaries
latm = 40.477399
latmm = 40.917577
lngm = -74.25909
lngmm = -73.700272
```

```python
n = 100
buckets = {}
```

```python
id_to_bin = {}
xx = []
x = range(n)
y = range(n)
for i in x:
    for j in y:
        xx.append((j*n)+i)
        id_to_bin[(j*n)+i] = (i,j)
```

```python
w = (lngmm - lngm)/float(n)
h = (latmm - latm)/float(n)
print w
print h
```

```python
#create lists of xticks and yticks
x = []
y = []
for i in xrange(0,n+1):
    x.append(lngm + (i*w))
    y.append(latm + (i*h))
```

```python
#c_d is a dictionary of centroids for buckets
scats = []
c_d = {}
xup = (x[1]-x[0])/2
yup = (y[1]-y[0])/2
for i in xrange(0,len(x)-1):
    for j in xrange(0,len(y)-1):
        scats.append((x[i]+xup,y[j]+yup))
        c_d[(i,j)] = (x[i]+xup,y[j]+yup)
```

```python
#given an id, convert it to bin and then convert to lat long centroid
id = 986
c_d[id_to_bin[id]]
```

```python
text_file = open('output.txt', "r") #100x100 grid, by 15 minute intervals
# lines = text_file.read().split(' ')
lines = text_file.readlines()
print len(lines)
text_file.close()
```

```python
# regroup all buckets
buckets = {}
for i in xrange(len(lines)):
    t = lines[i].split('|')
    bins = eval(t[1])[0]
    count = eval(t[1])[1]
    hr = t[0]
    
    if hr in buckets:
        if bins in buckets[hr]:
            buckets[hr][bins] += count
        else:
            buckets[hr][bins] = count
    else:
        buckets[hr] = {}
        buckets[hr][bins] = count 
```

```python
# create new dictionary with ids converted to lat longs
# keys are hours of the day each one being a dictionary that has lat,lngs as keys and counts as values
final = {}
for h in buckets:
    final[h] = {}
for h in buckets:
    for b in buckets[h]:
        final[h][c_d[id_to_bin[b]]] = buckets[h][b]
```

```python
import scipy.ndimage as ndi
```

```python
#create list of data items, one for each hour
data_list = []
```

```python
#create the grid to apply gaussian filter to
for i in xrange(24):
    for j in xrange(1,5):
        data = np.random.rand(n,n)
        data_list.append(data.astype(int))
```

```python
#populate the grid from the values in the buckets dictionary
hr_h = 0
hr_m = 1
ctr = 1
for data in data_list:
    hrcnt = str(hr_h)+'_'+str(hr_m)
    for i in xrange(n):
        for j in xrange(n):
            b = int(j*n + i)
            if b in buckets[hrcnt]:
                data[i,j] = buckets[hrcnt][b]
            else:
                data[i,j] = 0
    hr_m += 1
    if ctr % 4 == 0:
        hr_h += 1
        hr_m = 1
    ctr += 1
```

```python
img_list = []
for data in data_list:
    img = ndi.gaussian_filter(data, (3,3))  ## gaussian convolution
    img_list.append(img)
```

```python
filt_list = []
for img in img_list:
    filt = {}
    for i in xrange(n):
        for j in xrange(n):
            b = int(j*n + i)
            filt[b] = img[i,j]
    filt_list.append(filt)
```

```python
buckets_n = buckets
counter = 0
for i in xrange(24):
    for j in xrange(1,5):
        hr = str(i)+'_'+str(j)
        buckets_n[hr] = filt_list[counter]
        counter += 1
```

```python
final_n = {}
for h in buckets_n:
    final_n[h] = {}
for h in buckets_n:
    for b in buckets_n[h]:
        final_n[h][c_d[id_to_bin[b]]] = buckets_n[h][b]
```

## convert nys json into nyc json

```python
import json
```

```python
#read json as a string
json_file = open('nys/nys.json')
json_str = json_file.read()
```

```python
#convert to dictionary
json_data = json.loads(json_str)
```

```python
json_data['type']
```

```python
type(json_data['features'])
```

```python
#create dictionary for nyc
nyc_json = {}
nyc_json['type'] = u'FeatureCollection'
nyc_json['features'] = []
```

```python
#loop through nys dictionary and keep entries where city is New York
for i in json_data['features']:
    if i['properties']['LeftCityTo'] == 'New York':
        nyc_json['features'].append(i)
```
